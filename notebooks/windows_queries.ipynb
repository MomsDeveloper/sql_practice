```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mohammedalsubaie/amazon-products")
```

```python
import pandas as pd
```

```python
df = pd.read_pickle(path + "/Amazon_products.pkl")
df.head()
```

```python
df.info()
```

```python
df["rating"] = df["rating"].astype(str).apply(lambda x: x.replace(" out of 5 stars", ""))
```

```python
def solds_to_normal(x):
    return x.replace("K+", "000").replace("+", "")
```

```python
df["sold_past_month"] = df["sold_past_month"].astype(str).apply(solds_to_normal)
```

```python
def numbers_to_normal(x):
    return x.replace(",", "")
```

```python
df["price"] = df["price"].astype(str).apply(numbers_to_normal)
df["list_price"] = df["list_price"].astype(str).apply(numbers_to_normal)
df["reviews"] = df["reviews"].astype(str).apply(numbers_to_normal)
```

```python
float_cols = ["price", "list_price", "reviews", "rating", "sold_past_month", "available_offers"]
df[float_cols] = df[float_cols].astype(float)
```

```python
df[float_cols].describe()
```

#### Data Cleaning

```python
import os

# create directory if it does not exist
os.makedirs("../data", exist_ok=True)

# save df 
df.to_pickle("../data/Amazon_products.pkl")
```

#### 1. Getting rank of popularity and rating of each product in brand.

```python
df_sorted = df.sort_values(["rating", "reviews"], ascending=False)
df_sorted["rank"] = df_sorted.groupby("brand").cumcount() + 1
df_sorted.sort_values(["brand", "rank"]).head(5)
```

#### 2. Getting the best number of sold in month product in each brand.

```python
df_sorted = df.sort_values(["price"])
df_sorted["best_sold_in_month"] = df_sorted.groupby("brand")["sold_past_month"].transform("max")
df_sorted.sort_values(["best_sold_in_month"], ascending=False).head(5)
```

#### 3. Getting the rolling average of rating over the last three products (sorted by price) for each brand.

```python
df_sorted = df.sort_values(["brand", "price"])
df_sorted["avg_rating"] = df_sorted.groupby("brand")["rating"].transform(lambda x: round(x.rolling(3).mean(), 2))
df_sorted.head(5)
```

#### 4. Calculating the cumulative sum of available_offers for each is_prime group.

```python
df_sorted = df.sort_values(["is_prime"])

df_is_prime = df_sorted[df_sorted["is_prime"] == 1]
df_is_prime["cummulative_offer"] = df_is_prime["available_offers"].expanding().sum()

df_not_prime = df_sorted[df_sorted["is_prime"] == 0]
df_not_prime["cummulative_offer"] = 0

df_sorted = pd.concat([df_is_prime, df_not_prime]).sort_index()
df_sorted.sort_values(["is_prime"]).head()
```

#### 5. Within each brand, calculating the percentage rank of price for each product.

```python
df_sorted = df.sort_values(["brand", "price"])
df_sorted["max_price"] = df_sorted.groupby("brand")["price"].transform("max")
df_sorted["price_pecentage"] = df_sorted["price"] / df_sorted["max_price"]
df_sorted["price_rank"] = df_sorted.groupby("brand")["price_pecentage"].rank(ascending=False)
# df_sorted["price_rank"] = df_sorted.groupby("brand")["price_pecentage"].transform(lambda x: x.rank(ascending=False))
df_sorted.head(5)
```

#### 6. Compute the difference between the current amount of discount and the previous one for each brand

```python
df_sorted = df.copy()
df_sorted["price_discounted"] = df_sorted["list_price"] - df_sorted["price"]
df_sorted.sort_values(["brand", "price_discounted"], inplace=True)
df_sorted["price_diff_rank"] = df_sorted.groupby("brand")["price_discounted"].transform(lambda x: x - x.shift(1))
df_sorted.head(5)
```
