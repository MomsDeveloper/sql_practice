```python
import psycopg2
import pandas as pd

```

### Getting data from Amazon dataset to insert into an sql table 

```python
df = pd.read_pickle("../data/Amazon_products.pkl")
df.head()
```

### Creating new table with some columns from the dataset

```python
try:
    conn = psycopg2.connect(
        host="localhost",
        database="data",
        user="user",
        password="user")
    print("Cursor created")
    cur = conn.cursor()
    q = "DROP TABLE IF EXISTS Amazon_products"
    cur.execute(q)
    conn.commit()
    q = "CREATE TABLE IF NOT EXISTS Amazon_products (id SERIAL PRIMARY KEY, product_name VARCHAR(255), product_price FLOAT, product_rating FLOAT, product_reviews FLOAT)"

    cur.execute(q)
    conn.commit()
    print("Table created")
    for i in range(len(df)):
        q = f"INSERT INTO Amazon_products (product_name, product_price, product_rating, product_reviews) VALUES ('{df['asin'][i]}', '{df['price'][i]}', '{df['rating'][i]}', '{df['reviews'][i]}')"
        cur.execute(q)
        conn.commit()
    print("Data inserted")
    
    q = "SELECT * FROM Amazon_products"
    cur.execute(q)
    rows = cur.fetchall()
    for row in rows:
        print(row)
    
    cur.close()
    conn.close()
except Exception as e:
    print("Error: ", e)
```
